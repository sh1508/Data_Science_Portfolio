# Relationship Between Inflation and Interest Rates

__Inflation__ occurs when prices rise across the economy, decreasing the purchasing power of your money. Inflation refers to the broad increase in prices across a sector or an industry, like the automotive or energy business—and ultimately a country’s entire economy.

# Part 1: Understanding Inflation

## Measures of Inflation
The chief measures of U.S. inflation are the __Consumer Price Index (CPI)__, the __Producer Price Index (PPI)__ and the __Personal Consumption Expenditures Price Index (PCE)__, all of which use varying measures to track the change in prices consumers pay and producers receive in industries across the whole American economy.

### Consumer Price Index \(CPI):
The U.S. Bureau of Labor Statistics calculates CPI monthly based on the changes in prices consumers pay for goods and services. The CPI uses a “basket of goods” approach, meaning it tracks changes in the costs of eight major categories people spend money on: food and beverages, housing, apparel, transportation, education and communication, recreation, medical care, and other goods and services.

Many consider the CPI the benchmark for measuring inflation in the United States. The CPI is especially important because it is used to calculate cost of living increases for Social Security payments and for many companies’ annual raises. It is also used to adjust the rates on some inflation-protected securities, like Treasury Inflation-Protected Securities (TIPS).

### Personal Consumption Expenditures Price Index \(PCE):
Published by the Bureau of Economic Analysis, PCE tracks how much consumers pay for goods and services in the economy. This index considers a broader range of consumer expenditures than CPI, like healthcare spending. It also updates the basket of goods it uses for calculations based on what consumers are actually spending money on each month, rather than limiting data to a fixed set of goods.

PCE is an especially important because it’s the Federal Reserve’s preferred measure of inflation when making monetary decisions.

#### PCE vs. CPI: What's the Difference?
Generally, the CPI releases several weeks earlier than the PCE each month, while __the PCE is considered slightly more accurate__ than its counterpart. __The CPI usually shows higher inflation__ than the PCE due to differences in the formulas, components, and weights of each index. 

The CPI relies on a Laspeyres formula, whereas the PCE uses a Fisher-Ideal Formula. The PCE's Fisher-Ideal formula is considered more accurate as it accounts for __consumer substition__ of goods as relative prices change. For example, as the price of milk goes up, peiple frequently purchase less milk, which the PCE takes into account. However, the PCE formula depends on more current data, which explains why the PCE releases later than the CPI each month.

Each index also maintains different weightings for its categories of items. The CPI's weights are based on the Consumer Expenditure Survey, conducted by the Bureau of Labor Statistics, while the PCE's weights are based on spending data, like the Census Bureau's retail trade surveys. As a result, each inflation metric weighs some things slightly differently than the other. Famously, __housing makes up 33% of the CPI but just 15% of the PCE basket__.

The two indices use different scopes for measuring consumption. The CPI gauges purchases made by all urban households, while the PCE uses changes in goods and services consumed by all households, as well as nonprofit institutions serving households. As a result, some expenditures counted in the PCE are outside the CPI's scope. This contributes to the general opinion that the PCE is more inclusive and accurate than the CPI.


### Producer Price Index \(PPI):
Also published by the Bureau of Labor Statistics, PPI tracks the changes in prices that companies receive for the goods and services they sell each month. Costs can rise when producers face an increase in tariffs, higher oil and gas prices to transport their items, or other issues, such as the impact of a long-lasting pandemic or environmental changes, like a rise in hurricanes, wildfires, or flooding.

The PPI plays an important role in business contracts. Businesses that enter into long-term contracts with suppliers frequently use the PPI to automatically adjust the rate they pay for raw goods and services over time. Otherwise, suppliers would lock themselves into years-long contracts at rates that might lose them purchasing power over the long term.

### Core Inflation:
Core versions of CPI and PCE exclude food and evergy prices, which move less predictably because of weather conditions, geopolitical events, and seasonal changes. The Federal Reserve generally means core PCE when they discuss "underlying inflation."

## Is Inflation Good or Bad?
Though it can be frustrating to think about your dollars losing value, most economists consider a small amount of inflation a sign of a healthy economy. A moderate inflation rate encourages you to spend or invest your money today, rather than stuff it under your mattress and watch its value diminish.

__Inflation can become a destructive force in an economy if it is allowed to get out of hand and rise dramatically__. Unchecked inflation can topple a country’s economy, like in 2018 when Venezuela’s inflation rate hit over 1,000,000% a month, causing the economy to collapse and forcing countless citizens to flee the country.

### Deflation:
When prices decline across a sector of the economy or throughout the entire economy, it’s called deflation. While it might seem nice that you can buy more for less tomorrow, economists warn that deflation can be even more dangerous for an economy than unchecked inflation.

When deflation takes hold, consumers delay purchases in the present as they wait for prices to decline even further in the future. If left unchecked, deflation can diminish or freeze economic growth, which in turn decimates wages and paralyses an economy.

### Hyperinflation:
Hyperinflation occurs when inflation rises rapidly and the value of the currency of the country tumbles rapidly. Economists define hyperinflation as taking place when prices rise by at least 50% each month. Though rare, past instances of hyperinflation have taken place during civil unrest, during war time or when regimes have been taken over, rendering currency effectively worthless.

Perhaps the best-known example of hyperinflation took place in Weimar Germany, in the early 1920s. Prices rose by tens of thousands of percent each month, which very badly damaged the German economy.

### Stagflation:
Stagflation occurs when inflation remains high, but a country’s economy is not growing and its unemployment is rising. Usually, when unemployment increases, consumer demand decreases as people watch their spending more closely. This decrease in demand lowers prices, helping to recalibrate your purchasing power.

When stagflation happens, however, prices remain high even as consumer spending decreases, making it increasingly expensive to buy the same goods. We don’t have to look abroad to find examples, as the U.S. experienced stagflation in the mid to late 1970s, as high prices from OPEC oil embargoes drove inflation higher even as recession lowered GDP and increased unemployment.

## What Causes Inflation?
The gradually rising prices associated with inflation can be caused two main ways: __demand-pull__ inflation and __cost-push__ inflation. Both come back to the fundamental economic principles of supply and demand.

### Demand-Pull Inflation
Demand-pull inflation is when demand for goods or services increases but supply remains the same, pulling up prices.

Demand-pull inflation can be caused a few ways. In a healthy economy, people and companies increasingly make more money. This growing purchasing power allows consumers to buy more than they could before, increasing competition for existing goods and raising prices while companies attempt to ramp up production. On a smaller scale, demand-pull inflation can be caused by sudden popularity of certain products.

### Cost-Push Inflation
Cost-push inflation is when supply of goods or services is limited in some way but demand remains the same, pushing up prices. Usually, some sort of external event, like a natural disaster, hinders companies' abilities to produce enough of certain goods to keep up with consumer demand. This allows them to raise prices, resulting in inflation.

## How Interest Rates Affect Inflation
The Federal Reserve is the central bank of the U.S., and the Fed — like central banks around the world — is tasked with maintaining a stable rate of inflation. The Federal Open Markets Committee (FOMC) has determined that __an inflation rate around 2% is optimal employment and price stability__.

This level of inflation gives the FOMC scope to jump-start the economy during downturns by decreasing interest rates, which makes borrowing cheaper and helps boost consumption. Lower interest rates reduce costs for businesses and consumers to borrow money, stimulating the economy. Lower interest rates also mean individuals earn less on their savings, encouraging them to spend. But all this extra demand can push up inflations.

As prices rise, inflation effectively reduces the purchasing power of money because a dollar today just doesn't buy as much as it did, say, 10 years ago. In times of high inflation, the Fed raise interest rates to slow the pace of inflation by reducing overall demand and mitigating upward pressure on prices.

The initial response from consumers to rising interest rates is often a decreased willingness to finance major purchases such as homes, vehicles, and household appliances. At 3% mortgage rates, a family might be able to afford monthly payments for a \\$300,000 home, but 6% could make the same property unaffordable. When borrowing costs increase, businesses, too, tend to delay their growth plans and reduce capital expenditures, which can result in job losses and lower wage growth.

The Fed's primary tool for hitting this 2% target is __the Federal Funds Rate (FFR)__ — the interest rate banks charge each other for overnight loans.

The FOMC meets roughly every 6 weeks to set a target range for the FFR. To keep the effective FFR withiin the target range, the Fed uses three main tools:

### Interest on Reserve Balances (IORB):
This is the rate the Fed pays banks on their reserves held at the Fed. This acts as a floor as banks won't lend to other banks for less than what they could earn risk-free from the Fed.

### Overnight Reverse Repurchase Agreements (ON RRP):
The Fed borrows cash overnight from money market funds and others, offering Treasuries as collateral and paying interest. This tool helps absorb excess liquidity and strengthens the lower bound of FFR.

### Open Market Operations (OMOs):
The Fed buys and sells Treasury securities to adjust the amount of reserves in the banking system. This is less commonly used now on a day-to-day basis due to the large reserves environment where there is so much liquidity that OMOs aren't as effective for fine-tuning, but still important during monetary tightening or easing cycles.


When the Fed adjusts the FFR, the effects cascade through the rest of the economy:
- Banks immediately adjust their prime lending rates.
- Credit card companies update their adjustable percentage rates.
- New auto loans and adjustable-rate mortgages become more or less expensive as they are set a certain number of basis points above the overnight rate.
- Bond markets re-price fixed-income securities as newly issued bonds will reflect the new rates.
- Businesses reassess expansion plans based on new borrowing costs, which might also affect unemployment rate. <br><br>

# Part 2: Data Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CPI
Core_CPI = pd.read_csv('Price_Index/CPI/US_Core_CPI_by_Month_SA.csv', parse_dates=['Month'])
CPI = pd.read_csv('Price_Index/CPI/US_CPI_by_Month_SA.csv', parse_dates=['Month'])
Energy = pd.read_csv('Price_Index/CPI/US_Energy_CPI_by_Month_SNA.csv', parse_dates=['Month'])
Food = pd.read_csv('Price_Index/CPI/US_Food_CPI_by_Month_SNA.csv', parse_dates=['Month'])
# PCE
PCE = pd.read_csv('Price_Index/PCE/US_PCE_by_Month_SA.csv', parse_dates=['Date'])
# PPI
Core_PPI = pd.read_csv('Price_Index/PPI/US_Core_PPI_by_Month_SA.csv', parse_dates=['Month'])
PPI = pd.read_csv('Price_Index/PPI/US_PPI_by_Month_SA.csv', parse_dates=['Month'])
# FFR
FFR = pd.read_excel('US_Historical_FFR.xlsx' , parse_dates=['Date'])
# Real GDP
RGDP = pd.read_excel('US_Historical_Real_GDP_Growth Rate.xlsx' , parse_dates=['Date'])
# Unemployment Rate
UEMP = pd.read_excel('US_Historical_Unemployment_Rate.xlsx' , parse_dates=['Month'])
```

```python
# CPI
Core_CPI_slice = Core_CPI.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].rename(columns={'12-Month % Change': 'Core CPI YoY Change %'})
CPI_slice = CPI.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].rename(columns={'12-Month % Change': 'CPI YoY Change %'})
Energy_slice = Energy.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].rename(columns={'12-Month % Change': 'Energy YoY Change %'})
Food_slice = Food.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].rename(columns={'12-Month % Change': 'Food YoY Change %'})
# PCE
PCE_slice = PCE.set_index('Date').sort_index().loc['2000-07-01':'2025-04-30']
# PPI
Core_PPI_slice = Core_PPI.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].rename(columns={'12-Month % Change': 'Core PPI YoY Change %'})
PPI_slice = PPI.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].rename(columns={'12-Month % Change': 'PPI YoY Change %'})
# FFR
FFR_slice = FFR.set_index('Date').sort_index().loc['2000-07-01':'2025-04-30']
# Real GDP
RGDP_slice = RGDP.set_index('Date').sort_index().loc['2000-07-01':'2025-04-30'].astype({'YoY Change %': 'float'}).rename(columns={'YoY Change %': 'Real GDP YoY Change %'})
# Unemployment Rate
UEMP_slice = UEMP.set_index('Month').sort_index().loc['2000-07-01':'2025-04-30'].astype({'Unemployment Rate': 'float'})
```

```python
Core_PI = pd.concat([PCE_slice['Core PCE YoY Change %'], Core_CPI_slice['Core CPI YoY Change %'], Core_PPI_slice['Core PPI YoY Change %']], axis=1)
PI = pd.concat([PCE_slice['PCE YoY Change %'], CPI_slice['CPI YoY Change %'], PPI_slice['PPI YoY Change %']], axis=1)
```

## Historical Effective Federal Funds Rate (EFFR)

```python
colors = {
    'Rate (%)': '#9A0000',             # Dark Wine Red
    'Target Rate From (%)': '#ADADAD', # Gray
    'Target Rate To (%)': '#ADADAD'    # Gray
}

plt.figure(figsize=(12,6))

for col in FFR_slice.columns:
    if col in ['Target Rate From (%)', 'Target Rate To (%)']:
        plt.plot(
            FFR_slice.index,
            FFR_slice[col],
            label=col,
            color=colors[col],
            alpha=0.9,
            linestyle='--' 
        )
    else:
        plt.plot(
            FFR_slice.index,
            FFR_slice[col],
            label=col,
            color=colors[col],
            alpha=1.0,
            linestyle='-'   
        )

plt.text(FFR_slice.index[1050], 0.9, "1", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[1450], 5.3, "2", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[1700], 4.6, "3", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[2000], -0.1, "4", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[3920], -0.1, "5", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[4800], 2.6, "6", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[4860], -0.1, "7", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[5480], -0.1, "8", fontsize=9, fontstyle='italic')
plt.text(FFR_slice.index[6100], 5.5, "9", fontsize=9, fontstyle='italic')

plt.title("Historical Effective Federal Funds Rate", fontsize=14)

plt.legend();
```

The graph shows that the Fed has been successfully kept the effective FFR within the target range.

#### 1. Economic Expansion
The U.S. had faced a recession in the early 2000s due to the dot-com bust, along with the tragedy of 9/11 and conflicts like the Iraq War. But with a recorvering economy, the Fed raised rates for the first time in four years.


#### 2. A 17-Hike Streak
The Fed raised rates for the 17th time in a row, pushing the rate to 5.25% to curb inflation. The housing bubble was bursting.

#### 3. The Start of the Financial Crisis
As the housing market took a downturn, the Fed decided to cut interest rates to 4.75%. This move marked the beginning of a series of rate decreases.

#### 4. Fighting the Recession
In the thick of the Great Recession, the Fed slashed rates to a 0.00%-0.25% range — the lowest level in history — to help the US. economy cope with the fallout from the 2008 global financial crisis. 

#### 5. Returning to Normalcy
Seven years later, the central bank gingerly raised rates as the economy recovered gradually. “The Committee judges that there has been considerable improvement in labor market conditions this year, and it is reasonably confident that inflation will rise, over the medium term, to its 2% objective,” the Fed wrote in a statement that accompanied the initial 2015 rate hike.

#### 6. Mid-Cycle Adjustment
The Fed cut interest rates three times in 2019. The Fed eased rates midway through the typical expansion-to-recession business cycle. In 2019, the U.S. and China were in conflict over trade, and the Fed was concerned that the conflict would harm the economy and push up unemployment rates.

#### 7. Coping with Covid-19
The Fed delivered two huge cuts in response to the Covid-19 pandemic, returning the federal funds target rate range of zero to 0.25%. While the economy was technically growing again by May 2020, after the shortest recession on record, the fallout from the economic meaures to cope with the Covid-19 outbreak was still being felt.


#### 8. Taming Inflation
It wasn’t that long ago that the Fed was keeping the federal funds rate as low as possible as recently as the first few months of 2022. Remember, inflation was soaring for the better part of the year. But when the Fed realized that inflation was going to be much less transitory than it imagined, it shifted gears forcefully. Between March 2022 and July 2023, the Fed raised rates to 5.5% by July 2023.The tough medicine had the desired effect and inflation has dropped.

#### 9. Beyond
The unemployment rate began to drift upward. With the labor market weakening, the Fed changed course and cut the federal funds rate to a range of 4.75% to 5.00%. The Fed then had two more cuts down to a range of 4.25% to 4.50%. The Fed choose to maintainn this range at its meetings in Jan, Mar, and May 2025.<br><br>
